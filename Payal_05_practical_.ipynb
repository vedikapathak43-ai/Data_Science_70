```python
# Import libraries
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("online vs store shopping dataset.csv")

# Display first 5 rows
print("First 5 rows:")
print(df.head())

# Dataset info
print("\nDataset Info:")
print(df.info())

# Statistical summary
print("\nStatistical Summary:")
print(df.describe())

# -----------------------------------
# 🔹 FILTERING DATA
# -----------------------------------

# 1. Users who prefer Online shopping
online_users = df[df['shopping_preference'] == 'Online']
print("\nOnline Shopping Users:")
print(online_users.head())

# 2. High income users (income > 100000)
high_income = df[df['monthly_income'] > 100000]
print("\nHigh Income Users:")
print(high_income.head())

# 3. Users with high internet usage (> 6 hours)
high_internet = df[df['daily_internet_hours'] > 6]
print("\nHigh Internet Usage Users:")
print(high_internet.head())

# -----------------------------------
# 🔹 GROUPING DATA
# -----------------------------------

# Group by shopping preference and calculate mean
grouped_pref = df.groupby('shopping_preference').mean(numeric_only=True)
print("\nMean values by Shopping Preference:")
print(grouped_pref)

# Group by gender
grouped_gender = df.groupby('gender').mean(numeric_only=True)
print("\nMean values by Gender:")
print(grouped_gender)

# -----------------------------------
# 🔹 AGGREGATE STATISTICS
# -----------------------------------

# Aggregate functions
agg_stats = df.groupby('shopping_preference').agg({
    'monthly_income': ['mean', 'min', 'max'],
    'avg_online_spend': ['mean', 'min', 'max'],
    'avg_store_spend': ['mean', 'min', 'max']
})

print("\nAggregate Statistics:")
print(agg_stats)

# Count users per shopping preference
count_pref = df['shopping_preference'].value_counts()
print("\nCount of Shopping Preferences:")
print(count_pref)

# -----------------------------------
# 🔹 ADDITIONAL INSIGHTS
# -----------------------------------

# Which group spends more online?
max_online_spend = df.groupby('shopping_preference')['avg_online_spend'].mean().idxmax()
print("\nGroup with highest average online spend:", max_online_spend)

# Which city tier prefers online shopping most
city_pref = df.groupby('city_tier')['shopping_preference'].value_counts()
print("\nCity Tier vs Shopping Preference:")
print(city_pref)
```
