### Coin Toss Simulation

This code simulates a series of coin tosses and calculates the probability of getting heads or tails based on the simulation results.

```python
import numpy as np

# Number of coin tosses
num_tosses = 100

# Simulate coin tosses (0 for tails, 1 for heads)
# np.random.choice([0, 1], size=num_tosses) would also work
tosses = np.random.randint(0, 2, size=num_tosses)

# Count heads and tails
heads_count = np.sum(tosses == 1)
tails_count = np.sum(tosses == 0)

# Calculate probabilities
probability_heads = heads_count / num_tosses
probability_tails = tails_count / num_tosses

print(f"Total tosses: {num_tosses}")
print(f"Heads count: {heads_count}")
print(f"Tails count: {tails_count}")
print(f"Probability of Heads: {probability_heads:.2f}")
print(f"Probability of Tails: {probability_tails:.2f}")
```
