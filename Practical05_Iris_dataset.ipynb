```python
# ================================
# 1. Import Libraries
# ================================
import pandas as pd
import numpy as np
import zipfile

# ================================
# 2. Extract ZIP File
# ================================
with zipfile.ZipFile("iris.zip", 'r') as zip_ref:
    zip_ref.extractall()

# ================================
# 3. Load Correct File (iris.data)
# ================================
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

df = pd.read_csv("iris.data", header=None, names=columns)

print("Dataset Loaded Successfully ✅")
print("\nFirst 5 rows:")
print(df.head())

print("\nColumn Names:")
print(df.columns)

# ================================
# 4. Filtering Data
# ================================
filtered_data = df[df['sepal_length'] > df['sepal_length'].mean()]
print("\nFiltered Data (sepal_length > mean):")
print(filtered_data.head())

# ================================
# 5. Grouping Data
# ================================
grouped = df.groupby('species')
print("\nGrouped by species (Mean values):")
print(grouped.mean(numeric_only=True))

# ================================
# 6. Aggregate Functions
# ================================
print("\nSUM:")
print(df.sum(numeric_only=True))

print("\nMEAN:")
print(df.mean(numeric_only=True))

print("\nCOUNT:")
print(df.count())

print("\nMINIMUM:")
print(df.min(numeric_only=True))

print("\nMAXIMUM:")
print(df.max(numeric_only=True))

# ================================
# 7. All Aggregations Together
# ================================
print("\nAggregations grouped by species:")
print(df.groupby('species').agg(['sum','mean','count','min','max']))

print("\n✅ Practical Completed Successfully!")
```
