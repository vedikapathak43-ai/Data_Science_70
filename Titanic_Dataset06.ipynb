```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

sns.set_style("whitegrid")

# Load Dataset
df = pd.read_csv("/content/Titanic_dataset.csv")

print("\n==============================")
print("1. DATASET PREVIEW")
print("==============================\n")

display(
    df.head().style
    .set_properties(**{'text-align': 'center'})
    .set_table_styles([{'selector': 'th', 'props': [('text-align', 'center')]}])
)

print("\n==============================")
print("2. DATASET INFORMATION")
print("==============================\n")

df.info()

print("\n==============================")
print("3. MISSING VALUES BEFORE CLEANING")
print("==============================\n")

display(
    df.isnull().sum().to_frame(name="Missing Values").style
    .set_properties(**{'text-align': 'center'})
    .set_table_styles([{'selector': 'th', 'props': [('text-align', 'center')]}])
)

print("\n==============================")
print("4. DATA PREPROCESSING")
print("==============================\n")

df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df = df.drop(columns=['Cabin'])

print("Missing values handled successfully.")

print("\n==============================")
print("5. MISSING VALUES AFTER CLEANING")
print("==============================\n")

display(
    df.isnull().sum().to_frame(name="Missing Values").style
    .set_properties(**{'text-align': 'center'})
    .set_table_styles([{'selector': 'th', 'props': [('text-align', 'center')]}])
)

print("\n==============================")
print("6. SURVIVAL COUNT TABLE")
print("==============================\n")

display(
    df['Survived'].value_counts().to_frame(name="Count").style
    .set_properties(**{'text-align': 'center'})
    .set_table_styles([{'selector': 'th', 'props': [('text-align', 'center')]}])
)

print("\n==============================")
print("7. SURVIVAL COUNT GRAPH")
print("==============================\n")

plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")
plt.show()

print("\n==============================")
print("8. PASSENGER CLASS VS SURVIVAL")
print("==============================\n")

plt.figure(figsize=(6,4))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Passenger Class vs Survival")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()

print("\n==============================")
print("9. GENDER VS SURVIVAL")
print("==============================\n")

plt.figure(figsize=(6,4))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Gender vs Survival")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```
