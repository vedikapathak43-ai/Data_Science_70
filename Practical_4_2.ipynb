```python
import numpy as np

# Prompt the user to enter the dataset
data_input = input("Please enter your dataset values separated by commas (e.g., 1,2,3,4,5): ")

# Convert the input string to a list of numbers
try:
    # Split the input string by comma and convert each part to a float
    dataset = [float(x.strip()) for x in data_input.split(',')]

    if not dataset:
        print("No data entered. Please try again.")
    else:
        # Convert the list to a numpy array for statistical calculations
        data_array = np.array(dataset)

        # Compute the statistics
        mean_val = np.mean(data_array)
        median_val = np.median(data_array)
        variance_val = np.var(data_array)
        std_dev_val = np.std(data_array)

        # Display the results
        print(f"\nDataset: {dataset}")
        print(f"Mean: {mean_val:.4f}")
        print(f"Median: {median_val:.4f}")
        print(f"Variance: {variance_val:.4f}")
        print(f"Standard Deviation: {std_dev_val:.4f}")

except ValueError:
    print("Invalid input. Please ensure all values are numbers separated by commas.")
except Exception as e:
    print(f"An error occurred: {e}")
```
