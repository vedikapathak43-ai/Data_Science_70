An expert system is an AI program that simulates the knowledge and reasoning process of human experts to solve problems. This example demonstrates a simple rule-based expert system to recommend a drink based on weather conditions.

This simple example demonstrates decision making based on conditions, a core concept in expert systems. It recommends an activity based on the current weather.

```python
# Define the current weather conditions
weather = "sunny"
temperature = 25 # Celsius

# Decision-making logic (expert rules)
if weather == "sunny" and temperature > 20:
    recommendation = "Go to the beach!"
elif weather == "cloudy" and temperature > 15:
    recommendation = "Go for a walk in the park."
elif weather == "rainy":
    recommendation = "Stay indoors and read a book."
elif temperature < 10:
    recommendation = "Stay cozy inside with hot chocolate."
else:
    recommendation = "Enjoy a casual day."

print(f"Current weather: {weather}, Temperature: {temperature}째C")
print(f"Activity recommendation: {recommendation}")

print("\n--- Another example ---")
weather = "rainy"
temperature = 18

if weather == "sunny" and temperature > 20:
    recommendation = "Go to the beach!"
elif weather == "cloudy" and temperature > 15:
    recommendation = "Go for a walk in the park."
elif weather == "rainy":
    recommendation = "Stay indoors and read a book."
elif temperature < 10:
    recommendation = "Stay cozy inside with hot chocolate."
else:
    recommendation = "Enjoy a casual day."

print(f"Current weather: {weather}, Temperature: {temperature}째C")
print(f"Activity recommendation: {recommendation}")
```
