```python
import pandas as pd
import numpy as np
from IPython.display import display, HTML

# Function to center align tables
def display_centered(df):
    display(HTML("<div style='display:flex; justify-content:center'>" + df.to_html() + "</div>"))

# Load the dataset
df = pd.read_csv('ultimate_student_productivity_dataset_5000.csv')

print("\n========== DATASET PREVIEW ==========\n")
display_centered(df.head())

print("\n========== DATA FILTERING ==========\n")

# Selecting students with productivity score greater than 70
high_productivity_students = df[df['productivity_score'] > 70]

display_centered(high_productivity_students.head())


print("\n========== GROUPING AND AGGREGATE STATISTICS ==========\n")

# Group by academic level and calculate statistics
agg_stats = df.groupby('academic_level')[['study_hours', 'productivity_score', 'exam_score']].agg(['sum','mean','count','min','max'])

display_centered(agg_stats)


print("\n========== SAVING SUMMARY FILE ==========\n")

# Save results to CSV
agg_stats.to_csv('productivity_summary.csv')

print("File 'productivity_summary.csv' saved successfully.")
```
